## <font color=darkmagenta>Aim</font>

**This project aims to create the Decision tree Classifier on the 'Iris' dataset and visualize it graphically.**

## <font color=darkmagenta>Importing Libraries</font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## <font color=darkmagenta>Loading Datasets</font>

```python
dataset = pd.read_csv('iris.csv')
dataset.head()
```

```python
dataset.shape
```

```python
feature_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
```

## <font color=darkmagenta>Splitting Dataset</font>

```python
X = dataset.iloc[:,[ 1, 2, 3,4]].values
X
```

```python
y = dataset.iloc[:,[5]].values
y
```

## <font color=darkmagenta>Building the decision tree model</font>

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(X,y)
```

## <font color=darkmagenta>Visualizing with treeplot</font>

```python
from sklearn import tree
text_representation = tree.export_text(classifier)
print(text_representation)
```

```python
fig = plt.figure(figsize=(25,20))
treeplt = tree.plot_tree(classifier, feature_names = feature_names,
                       class_names=['Iris-setosa','Iris-versicolor', 'Iris-virginica'], filled=True)
treeplt
plt.show()
```
